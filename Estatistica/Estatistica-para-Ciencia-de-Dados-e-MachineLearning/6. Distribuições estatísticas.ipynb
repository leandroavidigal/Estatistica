## Variáveis contínuas

As variáveis contínuas são fundamentais no estudo da estatística e encontram-se em diversos campos de aplicação, como na engenharia, economia, ciências sociais e naturais. Neste capítulo, abordaremos as principais características e métodos para trabalhar com variáveis contínuas, utilizando Python no ambiente Jupyter Notebook.

#### O que são Variáveis Contínuas?

Variáveis contínuas são aquelas que podem assumir qualquer valor dentro de um intervalo contínuo. Diferentemente das variáveis discretas, que possuem valores específicos e distintos (como número de filhos ou quantidade de carros), as variáveis contínuas podem assumir uma infinidade de valores. Exemplos de variáveis contínuas incluem altura, peso, temperatura, tempo, e pressão.

#### Exemplos de Variáveis Contínuas

- **Altura**: Pode ser medida com precisão cada vez maior (por exemplo, 170.5 cm, 170.55 cm, 170.555 cm, etc.).
- **Peso**: Também pode ser medido com alta precisão (por exemplo, 65.2 kg, 65.23 kg, 65.234 kg, etc.).
- **Tempo**: Pode ser medido em qualquer fração de segundo (por exemplo, 10.5 segundos, 10.55 segundos, 10.555 segundos, etc.).

#### Importância das Variáveis Contínuas

Variáveis contínuas são essenciais na modelagem e análise de fenômenos naturais e sociais. Elas permitem uma descrição mais precisa e detalhada dos dados e ajudam a desenvolver modelos matemáticos que refletem melhor a realidade.

#### Trabalhando com Variáveis Contínuas em Python

Utilizando Python, podemos explorar variáveis contínuas de diversas maneiras, como através da geração de amostras, visualização de distribuições e cálculo de estatísticas. O ambiente Jupyter Notebook é particularmente útil para esses propósitos, pois permite uma interação dinâmica e visualização imediata dos resultados.

A seguir, apresentamos alguns conceitos e ferramentas essenciais para o estudo de variáveis contínuas:

1. **Distribuição de Probabilidade**: A função que descreve a probabilidade de uma variável contínua assumir um determinado valor.
2. **Função Densidade de Probabilidade (PDF)**: Descreve a densidade de probabilidade para valores contínuos, com a área sob a curva representando a probabilidade.
3. **Função de Distribuição Acumulada (CDF)**: Representa a probabilidade de que a variável aleatória seja menor ou igual a um determinado valor.
4. **Simulação e Amostragem**: Geração de dados sintéticos para modelagem e análise, utilizando funções disponíveis em bibliotecas como `numpy` e `scipy`.
5. **Visualização de Dados**: Ferramentas para plotar distribuições e dados contínuos, utilizando bibliotecas como `matplotlib` e `seaborn`.

### Distribuição normal - Gaussian distribution (bell)

A distribuição normal, também conhecida como distribuição gaussiana ou curva em formato de sino (bell curve), é uma das distribuições de probabilidade mais importantes e amplamente utilizadas na estatística. Esta distribuição desempenha um papel central na teoria das probabilidades e na estatística inferencial devido às suas propriedades matemáticas e à sua ubiquidade em fenômenos naturais e sociais.

#### O que é a Distribuição Normal?

A distribuição normal é uma distribuição contínua de probabilidade que é simétrica em torno da média. Seu gráfico, conhecido como a curva de sino, tem uma forma característica onde a maioria dos valores se concentra em torno da média, e as probabilidades diminuem à medida que nos afastamos dela. Matematicamente, a função densidade de probabilidade (PDF) de uma distribuição normal é definida por:

$$f(x | \mu, \sigma) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}$$

Onde:
- $\mu$ é a média da distribuição,
- $\sigma$ é o desvio padrão,
- $\sigma^2$ é a variância,
- $e$ é a base do logaritmo natural (aproximadamente 2.718),
- $\pi$ é a constante pi (aproximadamente 3.14159).

#### Características da Distribuição Normal

1. **Simetria**: A distribuição é perfeitamente simétrica em torno da média (\( \mu \)).
2. **Média, Mediana e Moda**: Na distribuição normal, a média, a mediana e a moda são iguais e localizam-se no centro da distribuição.
3. **Curva em Sino**: O formato da curva é tal que os valores mais próximos da média têm maior probabilidade de ocorrência.
4. **Empirical Rule (Regra Empírica)**: Aproximadamente 68% dos dados estão dentro de um desvio padrão (\( \sigma \)) da média, 95% estão dentro de dois desvios padrão, e 99.7% estão dentro de três desvios padrão.

#### Importância da Distribuição Normal

A distribuição normal é fundamental em estatística por várias razões:

- **Teorema Central do Limite**: Este teorema afirma que, independentemente da distribuição original da população, a distribuição das médias amostrais tende a ser normal se o tamanho da amostra for suficientemente grande.
- **Modelagem de Fenômenos Naturais**: Muitos fenômenos naturais e sociais, como altura, peso, pontuações de testes e erros de medição, seguem aproximadamente uma distribuição normal.
- **Facilidade de Cálculo**: A simetria e as propriedades matemáticas da distribuição normal facilitam a análise e a inferência estatística.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Criar a distribuição normal
mean = 0
std_dev = 1
x = np.linspace(-3*std_dev, 3*std_dev, 1000)
y = (1/(std_dev * np.sqrt(2 * np.pi))) * np.exp(-(x - mean)**2 / (2 * std_dev**2))

# Plotar a distribuição normal
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue')

# Preencher as áreas sob a curva
plt.fill_between(x, y, where=(x >= -std_dev) & (x <= std_dev), color='darkblue', alpha=0.6)
plt.fill_between(x, y, where=(x >= -2*std_dev) & (x <= -std_dev), color='blue', alpha=0.6)
plt.fill_between(x, y, where=(x >= std_dev) & (x <= 2*std_dev), color='blue', alpha=0.6)
plt.fill_between(x, y, where=(x >= -3*std_dev) & (x <= -2*std_dev), color='lightblue', alpha=0.6)
plt.fill_between(x, y, where=(x >= 2*std_dev) & (x <= 3*std_dev), color='lightblue', alpha=0.6)

# Adicionar as porcentagens
plt.text(0, 0.1, '34.1%', ha='center', fontsize=12, color='white')
plt.text(-1.5*std_dev, 0.02, '13.6%', ha='center', fontsize=12)
plt.text(1.5*std_dev, 0.02, '13.6%', ha='center', fontsize=12)
plt.text(-2.5*std_dev, 0.005, '2.1%', ha='center', fontsize=12)
plt.text(2.5*std_dev, 0.005, '2.1%', ha='center', fontsize=12)
plt.text(-3*std_dev, 0, '0.1%', ha='center', fontsize=12)
plt.text(3*std_dev, 0, '0.1%', ha='center', fontsize=12)

# Adicionar rótulos dos eixos e título
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Distribuição Normal (Distribuição Gaussiana)')
plt.grid(True)
plt.show()
```

O gráfico apresentado é uma representação da **distribuição normal**, também conhecida como **distribuição gaussiana** ou **curva de sino**. Esta distribuição é fundamental em estatística devido à sua frequência em fenômenos naturais e propriedades matemáticas úteis.

#### Componentes do Gráfico

1. **Eixo Horizontal (x)**:
   - Representa os valores da variável aleatória que segue uma distribuição normal.
   - Os valores variam de -3 desvios padrão (\(\sigma\)) a +3 desvios padrão (\(\sigma\)) da média (\(\mu\)), que está centralizada em 0.

2. **Eixo Vertical (f(x))**:
   - Representa a densidade de probabilidade dos valores.
   - A altura da curva em qualquer ponto fornece a probabilidade relativa de ocorrer aquele valor.

3. **Curva em Sino**:
   - A forma característica da curva de sino é simétrica em torno da média.
   - A curva atinge seu pico na média, e a probabilidade diminui à medida que você se afasta da média em ambas as direções.

#### Áreas Sob a Curva

A distribuição normal tem propriedades específicas que ajudam a entender a probabilidade dos dados caírem dentro de certos intervalos:

1. **34.1% dentro de \(-1\sigma\) a \(1\sigma\)**:
   - A área mais escura no centro do gráfico representa os valores que estão dentro de um desvio padrão da média.
   - Aproximadamente 68.2% dos dados caem dentro de \(\pm 1\sigma\) (34.1% em cada lado da média).

2. **13.6% dentro de \(-2\sigma\) a \(-1\sigma\) e de \(1\sigma\) a \(2\sigma\)**:
   - As áreas azuis adjacentes à área central representam os valores que estão entre um e dois desvios padrão da média.
   - Aproximadamente 27.2% dos dados caem dentro de \(\pm 2\sigma\) (13.6% em cada lado fora do primeiro desvio padrão).

3. **2.1% dentro de \(-3\sigma\) a \(-2\sigma\) e de \(2\sigma\) a \(3\sigma\)**:
   - As áreas mais claras na extremidade representam os valores que estão entre dois e três desvios padrão da média.
   - Aproximadamente 4.2% dos dados caem dentro de \(\pm 3\sigma\) (2.1% em cada lado fora dos dois primeiros desvios padrão).

4. **0.1% além de \(\pm 3\sigma\)**:
   - As áreas extremamente pequenas nas pontas representam os valores que estão além de três desvios padrão da média.
   - Apenas cerca de 0.2% dos dados caem fora de \(\pm 3\sigma\) (0.1% em cada lado).

#### Importância das Áreas

- **Regra Empírica (68-95-99.7)**:
  - Cerca de 68% dos dados estão dentro de \(\pm 1\sigma\).
  - Cerca de 95% dos dados estão dentro de \(\pm 2\sigma\).
  - Cerca de 99.7% dos dados estão dentro de \(\pm 3\sigma\).

Esta regra é muito útil para fazer inferências rápidas sobre a distribuição dos dados e para identificar outliers.

#### Exemplos de Aplicação

- **Altura das Pessoas**: A maioria das pessoas tem altura próxima da média, com poucas pessoas muito altas ou muito baixas.
- **Notas de Exames**: Em um grande grupo de estudantes, a maioria das notas estará próxima da média, com poucas notas muito altas ou muito baixas.

### Entendendo o Enviesamento (Skewness) em Distribuições

Enviesamento, ou skewness, é uma medida da assimetria na distribuição de dados. Ele indica a direção e o grau de desvio da distribuição em relação à distribuição normal (simétrica). Existem três tipos principais de enviesamento:

1. **Distribuição Normal (Simétrica)**
2. **Enviesamento Positivo (Skewed Right)**
3. **Enviesamento Negativo (Skewed Left)**

Vamos explorar cada tipo de enviesamento e gerar os gráficos correspondentes usando Python.

#### 1. Distribuição Normal (Simétrica)

Na distribuição normal, os dados são distribuídos simetricamente em torno da média. A curva em forma de sino é perfeitamente simétrica, com a média, a mediana e a moda coincidentes no centro da distribuição.

```python
# Gerar dados de uma distribuição normal
data_normal = np.random.normal(0, 1, 1000)

# Plotar a distribuição normal
plt.figure(figsize=(6, 4))
sns.histplot(data_normal, kde=True, bins=30, color='blue')
plt.title('Distribuição Normal')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()
```

#### 2. Enviesamento Positivo (Skewed Right)

Na distribuição com enviesamento positivo, a cauda da distribuição se estende mais para a direita. Isso significa que há uma concentração maior de valores menores e alguns valores muito altos puxam a cauda para a direita. A média é maior que a mediana.

```python
# Gerar dados com enviesamento positivo
data_pos_skew = np.random.exponential(scale=1, size=1000)

# Plotar a distribuição com enviesamento positivo
plt.figure(figsize=(6, 4))
sns.histplot(data_pos_skew, kde=True, bins=30, color='green')
plt.title('Enviesamento Positivo')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()
```

#### 3. Enviesamento Negativo (Skewed Left)

Na distribuição com enviesamento negativo, a cauda da distribuição se estende mais para a esquerda. Isso significa que há uma concentração maior de valores mais altos e alguns valores muito baixos puxam a cauda para a esquerda. A média é menor que a mediana.

```python
# Função para gerar dados com enviesamento negativo
def generate_neg_skewed_data(size):
    data = np.random.beta(a=2, b=5, size=size)  # Utilizando a distribuição Beta
    data = 3 - data  # Invertendo a distribuição para enviesamento negativo
    return data

# Gerar dados com enviesamento negativo
data_neg_skew = generate_neg_skewed_data(1000)

# Plotar a distribuição com enviesamento negativo
plt.figure(figsize=(6, 4))
sns.histplot(data_neg_skew, kde=True, bins=30, color='red')
plt.title('Enviesamento Negativo')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()
```

## Padronização (z-score) e k-NN

#### k-Nearest Neighbors (kNN)

O k-Nearest Neighbors (kNN) é um dos algoritmos de aprendizado de máquina mais simples e intuitivos. Ele é frequentemente utilizado para problemas de classificação e regressão e baseia-se na ideia de que objetos similares estão próximos uns dos outros em um espaço de características.

#### Como Funciona o kNN?

O kNN é um algoritmo baseado em instâncias, o que significa que ele não constrói um modelo explícito a partir do conjunto de treinamento. Em vez disso, ele armazena os casos de treinamento e faz a predição para novos casos com base nas distâncias dos casos de treinamento armazenados. Aqui está o passo a passo do funcionamento do kNN:

1. **Escolha do valor de \( k \)**: O primeiro passo é escolher o número de vizinhos (\( k \)). Este é um hiperparâmetro que precisa ser definido pelo usuário. A escolha de \( k \) pode influenciar significativamente o desempenho do algoritmo.

2. **Cálculo da distância**: Para cada ponto de dados no conjunto de teste, o algoritmo calcula a distância entre esse ponto e todos os pontos do conjunto de treinamento. A distância mais comum usada é a distância euclidiana, definida pela fórmula:

   $$
   DE(x, y) = \sqrt{\sum_{i=1}^{p} (x_i - y_i)^2}
   $$

   Onde:
   - $x$ e $y$ são os vetores de características dos pontos de dados.
   - $p$ é o número de características.

3. **Identificação dos k vizinhos mais próximos**: Após calcular as distâncias, os \( k \) pontos de dados do conjunto de treinamento que são mais próximos do ponto de teste são selecionados.

4. **Classificação**: No caso de problemas de classificação, o ponto de teste é classificado com base na maioria das classes de seus \( k \) vizinhos mais próximos. No caso de problemas de regressão, a média dos valores dos \( k \) vizinhos mais próximos é utilizada como predição.

#### Exemplo Visual

Vamos utilizar um exemplo visual para ilustrar o funcionamento do kNN. Suponha que temos um conjunto de pontos de dados classificados como vermelhos e azuis e queremos determinar a classe de um novo ponto de dados (representado pelo ponto de interrogação).

- **K = 1**: O vizinho mais próximo é azul, então o ponto é classificado como azul.
- **K = 2**: Considerando os dois vizinhos mais próximos (um azul e um vermelho), a classificação pode ser ambígua e pode ser resolvida com desempate.
- **K = 3**: Dos três vizinhos mais próximos, dois são vermelhos e um é azul, então o ponto é classificado como vermelho.
- **K = 4**: Dos quatro vizinhos mais próximos, dois são vermelhos e dois são azuis. Novamente, pode ser necessário desempate.
- **K = 7**: A maioria dos sete vizinhos mais próximos é vermelha, então o ponto é classificado como vermelho.

#### Considerações Importantes

- **Escolha de \( k \)**: A escolha do valor de \( k \) é crucial. Valores muito pequenos podem levar a um modelo ruidoso, enquanto valores muito grandes podem suavizar demais a fronteira de decisão.
- **Escalonamento dos Dados**: É importante que os dados sejam escalonados (normalizados) antes de aplicar o kNN, especialmente quando as características têm diferentes escalas.
- **Complexidade Computacional**: O kNN pode ser computacionalmente intensivo para grandes conjuntos de dados, pois a distância precisa ser calculada entre o ponto de teste e todos os pontos de treinamento.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Gerar dados de exemplo
np.random.seed(0)
blue_points = np.random.rand(10, 2) * 0.6
red_points = np.random.rand(10, 2) * 0.6 + 0.4
X = np.vstack((blue_points, red_points))
y = np.array([0] * 10 + [1] * 10)

# Novo ponto a ser classificado
new_point = np.array([[0.5, 0.5]])

# Configuração do classificador kNN
k = 3
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X, y)
prediction = knn.predict(new_point)

# Plotar os pontos de dados
plt.figure(figsize=(8, 6))
plt.scatter(blue_points[:, 0], blue_points[:, 1], color='blue', label='Classe Azul')
plt.scatter(red_points[:, 0], red_points[:, 1], color='red', label='Classe Vermelha')
plt.scatter(new_point[0, 0], new_point[0, 1], color='green', label='Novo Ponto (?)', edgecolor='black', s=100)

# Plotar as linhas conectando os vizinhos mais próximos
neighbors = knn.kneighbors(new_point, return_distance=False)
for neighbor in neighbors[0]:
    plt.plot([new_point[0, 0], X[neighbor, 0]], [new_point[0, 1], X[neighbor, 1]], 'k--', lw=0.5)

# Adicionar legenda e título
plt.legend()
plt.title(f'k-Nearest Neighbors (k = {k})')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.grid(True)
plt.show()
```

### Dados Enviesados e Como o Machine Learning Ajuda

#### O que são Dados Enviesados?

Dados enviesados, ou skewed data, são conjuntos de dados onde a distribuição não é simétrica. Em outras palavras, os dados são distorcidos em uma direção específica, resultando em uma cauda mais longa em uma extremidade da distribuição. Existem dois tipos principais de enviesamento:

1. **Enviesamento Positivo (Skewed Right)**:
   - A cauda da distribuição se estende mais para a direita.
   - Exemplos: Renda (algumas pessoas ganham muito mais do que a maioria), preços de imóveis.

2. **Enviesamento Negativo (Skewed Left)**:
   - A cauda da distribuição se estende mais para a esquerda.
   - Exemplos: Idade de aposentadoria (a maioria das pessoas se aposenta em uma idade similar, mas alguns se aposentam muito mais cedo), notas de exames (alguns alunos podem tirar notas muito baixas).

#### Problemas Causados por Dados Enviesados

Dados enviesados podem criar vários problemas na análise de dados e no desenvolvimento de modelos de machine learning:

1. **Medições Enganosas**:
   - As medidas de tendência central (média, mediana, moda) podem não representar adequadamente a "média" esperada dos dados.

2. **Inferência Errada**:
   - Pode levar a conclusões erradas sobre a população ou fenômeno estudado.

3. **Desempenho do Modelo**:
   - Modelos de machine learning podem ter desempenho prejudicado, pois podem ser tendenciosos em favor da classe dominante ou valores extremos.

#### Como o Machine Learning Ajuda a Lidar com Dados Enviesados

Existem várias técnicas e práticas em machine learning que ajudam a lidar com dados enviesados:

1. **Transformações de Dados**:
   - Aplicar transformações matemáticas aos dados pode ajudar a reduzir o enviesamento. Exemplos incluem transformações logarítmicas, raízes quadradas ou Box-Cox.

2. **Sobreamostragem e Subamostragem**:
   - **Sobreamostragem**: Aumentar a quantidade de exemplos da classe minoritária.
   - **Subamostragem**: Reduzir a quantidade de exemplos da classe majoritária.
   - Técnicas como SMOTE (Synthetic Minority Over-sampling Technique) são frequentemente usadas para gerar novos exemplos sintéticos da classe minoritária.

3. **Métricas de Avaliação Apropriadas**:
   - Utilizar métricas de avaliação que são apropriadas para dados enviesados, como F1-score, precisão e recall, ao invés de apenas acurácia.

4. **Algoritmos Específicos**:
   - Alguns algoritmos de machine learning são mais robustos a dados enviesados, como árvores de decisão e Random Forest.

5. **Balanceamento de Classes**:
   - Ajustar os pesos das classes no modelo de aprendizado para que o algoritmo dê mais importância à classe minoritária durante o treinamento.

```python
### Exemplo Prático: Transformação Logarítmica

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Gerar dados enviesados positivos
data_pos_skew = np.random.exponential(scale=1, size=1000)

# Aplicar transformação logarítmica
data_log_transformed = np.log(data_pos_skew + 1)

# Plotar os dados antes e depois da transformação
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.histplot(data_pos_skew, kde=True, bins=30, ax=axes[0], color='green')
axes[0].set_title('Dados Enviesados Positivos')

sns.histplot(data_log_transformed, kde=True, bins=30, ax=axes[1], color='blue')
axes[1].set_title('Dados Após Transformação Logarítmica')

plt.show()
```

## Inicialização de pesos em redes neurais

A inicialização de pesos é um passo crucial no treinamento de redes neurais. A maneira como os pesos são inicializados pode afetar significativamente a velocidade de convergência do algoritmo de aprendizado e a capacidade da rede de encontrar um bom mínimo na função de perda.

#### Por que a Inicialização de Pesos é Importante?
- Convergência Rápida: Uma boa inicialização ajuda a rede a convergir mais rapidamente durante o treinamento.
- Evitar Explosão ou Desaparecimento de Gradientes: Pesos mal inicializados podem levar a problemas de explosão ou desaparecimento de gradientes, dificultando o aprendizado da rede.
- Quebra da Simetria: Inicializar todos os pesos com o mesmo valor impede que diferentes neurônios aprendam características diferentes dos dados. Inicializações aleatórias quebram essa simetria, permitindo que neurônios diferentes aprendam funções diferentes.

### Teste de Shapiro-Wilk

- p-value é usado para interpretar o teste estatístico
- p <= alpha: rejeita a hipótese, não é normal
- p > alpha: não rejeita a hipótese, é normal

```python
from scipy.stats import shapiro
```

```python
_, p = shapiro(dados_normais)
p
```

```python
alpha = 0.05
if p > alpha:
  print('Distribuição normal')
else:
  print('Distribuição não normal')
```

```python
_, p = shapiro(dados_nao_normais)
p
```

```python
alpha = 0.05
if p > alpha:
  print('Distribuição normal')
else:
  print('Distribuição não normal')
```
